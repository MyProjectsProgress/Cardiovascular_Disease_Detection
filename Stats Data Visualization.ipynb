```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cardio_Cleaned_Data.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop("Unnamed: 0", inplace=True, axis=1)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data = df)
plt.xticks(rotation=40)
plt.show()
```

# Lets know the smoking percentage between males and females.

```python
df['gender']=df['gender'].replace(1,'Female')
df['gender']=df['gender'].replace(2,'Male')
df["gender"].unique()
```

```python
plt.figure(figsize=(7,7))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="gender", y="smoke", data=df, palette="Set2")
plt.title("Relation Between Gender and Smoking Precentage", fontsize = 14)
plt.xlabel("Gender", fontsize = 14)
plt.ylabel("Smoking Percentage", fontsize = 14)
plt.show()
```

# Lets know the Alcohol percentage between males and females.

```python
plt.figure(figsize=(7,7))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="gender", y="alco", data=df, palette="Set2")
plt.title("Relation Between Gender and Drinking Alcohol Precentage", fontsize = 14)
plt.xlabel("Gender", fontsize = 14)
plt.ylabel("Alcohol Percentage", fontsize = 14)
plt.show()
```

# Now time for comparing between males and females.

## Frist: Activity Index

```python
plt.figure(figsize=(7,7))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="gender", y="active", data=df, palette="Blues_d")
plt.title("Relation Between Gender and Activity", fontsize=14)
plt.xlabel("Gender", fontsize = 14)
plt.ylabel("Activity Index", fontsize = 14)
plt.show()
```

## Second: Weight

```python
plt.figure(figsize=(7,7))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="gender", y="weight", data=df, palette="pastel")
plt.title("Relation Between Gender and Weight", fontsize = 14)
plt.xlabel("Gender", fontsize = 14)
plt.ylabel("Weight", fontsize = 14)
plt.show()
```

## Third: Cardiovascular Disease

```python
plt.figure(figsize=(7,7))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="gender", y="cardio", data=df, palette="flare")
plt.title("Relation Between Gender and Cardiovascuar Disease", fontsize=14)
plt.xlabel("Gender", fontsize = 14)
plt.ylabel("Cardiovascular Disease", fontsize = 14)
plt.show()
# Actually both male and female have the same avivity index
```

# Number of smokers and non-smokers in the dataset

```python
df['smoke']=df['smoke'].replace(1,'Smoker')
df['smoke']=df['smoke'].replace(0,'Not Smoker')
df["smoke"].unique()
```

```python
plt.figure(figsize=(7,7))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="smoke", data=df)
plt.title("The Numbers of Smokers and Non-Smokers", fontsize = 14)
plt.xlabel("Smoking", fontsize = 14)
plt.ylabel("Number of People", fontsize = 14)
plt.show()
# After seeing the weak correlation between smoke and cardiovascular disease I felt something is not okay but after 
# knowing that less than 10% smokes in this dataset I understood.
```

```python
# Data is evenly splitted 
plt.figure(figsize=(7,7))
plt.figure(figsize=(10,5))
sns.countplot(df["cardio"])
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x = "cholesterol" , y = "ap_hi" , data = df, hue="gender",palette="husl")
#plt.legend(["Total Bill", "Size"] , title = "Scatter Plot")
plt.xlabel("Cholesterol Level", fontsize = 14)
plt.ylabel("High Blood Pressure", fontsize = 14)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x="age", hue="cardio", data=df, palette="colorblind", 
              edgecolor= sns.color_palette("dark", n_colors=1))
plt.title("The Relationship Between Cardio Vascular Disease and Age", fontsize = 14)
plt.xlabel("Age", fontsize = 14)
plt.ylabel("Number of People", fontsize = 14)
plt.show()
```
